Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

1. Generating 1000 random (x,y) coordinates and plotting the points

```python
x_data=np.random.normal(1,0.4,1000)
y_data=np.random.normal(2,1.2,1000)
plt.scatter(x_data,y_data,color="navy")
```

2. Simple Linear Regression from scratch

```python
def linear_regression(x,y):
    slope = (y*x).sum() - ((y.sum()*x.sum())/1000) / (x*x).sum() - ((x.sum()*x.sum())/1000)
    intercept = y.mean() - slope * x.mean()
    plt.plot(x,slope*x+intercept,color="green")
    plt.scatter(x,y,color="navy")
    print("slope: ",slope)
    print("intercept: ",intercept)
```

```python
linear_regression(x_data,y_data)
```

3. Implementing Gradient Descent to find Slope and Intercept values using average sum of squared error

```python
def gradient_descent(x,y):
    slope=0
    intercept=0
    learning_rate=0.5
    for i in range(1000): 
        y_pred=slope*x + intercept 
        derived_slope = (-2/1000)* x*(y-y_pred).sum()  
        derived_intercept = (-2/1000) * (y-y_pred).sum() 
        slope = slope - learning_rate* derived_slope  
        intercept = intercept - learning_rate * derived_intercept
    plt.plot(x,slope*x+intercept,color="green")
    plt.scatter(x,y,color="navy")
    print("slope: ",slope)
    print("intercept: ",intercept)
```

```python
gradient_descent(x_data,y_data)
```

4. Linear Regression Using Sklearn module

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.20,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
y_pred = reg.predict(x_test.reshape(-1,1))
```

```python
plt.scatter(x_train,y_train,color="navy")
plt.plot(x_test,y_pred,color="green")
```

Conclusion: After comparing slope and intercept values for the 3 different ways of obtaining them, it can be observed that gradient descent is the most accurate form of calculation. Moreover, using Scikit learn over building a model from scratch leads to better prediction.
